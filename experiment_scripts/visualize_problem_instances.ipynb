```python
from pathlib import Path
import os
import sys

main_directory = Path(os.path.abspath("")).parent
sys.path.append(str(main_directory))
```

```python
from experiment_scripts.benchmark_utility import load_benchmarking_dataset
dataset = load_benchmarking_dataset()
```

```python
from queasars.job_shop_scheduling.visualization import plot_jssp_problem_instance_gantt

n_qubits = 15
instance_nr = 5

print("minimum makespan: ", dataset[n_qubits][instance_nr][1])
plot = plot_jssp_problem_instance_gantt(problem_instance=dataset[n_qubits][instance_nr][0])
```

```python
def plot_instances(instances: list[tuple[int, int]], name_prefix: str) -> None:
    for i, (qubits, instance_nr) in enumerate(instances):
        filename = Path(main_directory, f"{name_prefix}_{i}.pdf")
        print((qubits, instance_nr), ": ", dataset[qubits][instance_nr][1])
        plot_jssp_problem_instance_gantt(problem_instance=dataset[qubits][instance_nr][0], colormap="Paired", save_path=filename)
```

```python
plot_instances([(21, 6), (21, 7), (21, 8), (21, 9), (21, 10)], name_prefix="benchinstance_21qubits")
```
